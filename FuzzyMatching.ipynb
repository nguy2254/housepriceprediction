```python
def match_name(name, list_names, min_score=0):
    #-1 score incase we dont' get any matches 
    max_score = -1 
    #return empty name for no match as well 
    max_name = ""
    # Iternating over all names in the others 
    for name2 in list_names:
        #finnd fuzzy match score 
        score = fuzz.ratio(name, name2)
        #check if we are above the threshold and have a better score
        if (score> min_score) & (score > max_score):
            max_name = name2
            max_score = score 
    return (max_name, max_score)
```
